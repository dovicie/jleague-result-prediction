```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df_toto_2020 = pd.read_csv("./toto_data/2020.csv", header=[0,1], index_col=[0,1])
df_toto_2020.rename(columns={'Unnamed: 2_level_1':'','Unnamed: 3_level_1':''}, level=1, inplace=True)
df_toto_2020
```

```python
df_pred_2020 = pd.read_csv("./predict_proba_data/2020.csv", index_col=0)
df_pred_2020
```

```python
df_price =  pd.DataFrame([[100,300,900,2700,8100,24300],[200,600,1800,5400,16200],[400,1200,3600,10800],[800,2400,7200],[1600,4800],[3200]], columns=['t_0','t_1','t_2','t_3','t_4','t_5'], index = ['d_0','d_1','d_2','d_3','d_4','d_5'])
df_price
```

```python
total_amount = 0
total_profit = 0
total_winning = 0

for index,row in df_toto_2020.iterrows():
    
    amount = 0
    d_count = 0
    t_count = 0
    
    buy_lots = []
    
    for i in range(1,6):
        m_id = row[(f"Match{i}","MatchID")] # これを引数で関数化する
        pred = (df_pred_2020.loc[m_id])
        
        # 買い方を返す
        buy_lot = []

        
        if pred[1] - pred[2] > 0.15:
            buy_lot = [1]
        elif pred[1] - pred[2] > 0.05:
            buy_lot = [0,1]
            d_count += 1
        elif pred[1] - pred[2] > -0.05:
            buy_lot = [0,1,2]
            t_count += 1
        elif pred[1] - pred[2] > -0.15:
            buy_lot = [0,2]
            d_count += 1
        else:
            buy_lot = [2]
            
        buy_lots.append(buy_lot)
        
        
    amount = int(df_price.iloc[d_count,t_count])
#     print(f'第{index[0]}回 {index[1]}組  購入金額 {amount}円  買うくじ {buy_lots} ')
    
    # 実際に買っていたときの結果を出力
    winning = row[("Winning","")]
    is_winning = True
    for i in range(1,6):
        m_wl = row[(f"Match{i}","W/L")]
        if not m_wl in buy_lots[i-1]:
            is_winning = False
            
    if is_winning:
        profit = winning - amount
        total_winning += winning
    else:
        profit = amount * (-1)
        
#     print(f"リターン {profit}円")
    
    total_amount += amount
    total_profit += profit
    pay_back_rate = total_winning / total_amount
    
print(f"総購入金額{total_amount}円 総利益{total_profit}円 還元率{pay_back_rate:.02}")   
```

### predict_probaを学習させる

```python
df_pred_2020 = pd.read_csv("./predict_proba_data/2020.csv", index_col=0)
df_pred_2020.sort_index(inplace=True)
df_pred_2020
```

```python
df_2020 = pd.read_csv("./match_data_yearly/2020.csv", index_col="ID")
df_2020.sort_index(inplace=True)
df_2020
```

```python
df_2020["W/L"]
```

```python
df = pd.concat([df_pred_2020, df_2020["W/L"]], axis=1)
df
```

```python
X_train = df[["0","1","2"]].values[:270]
X_test = df[["0","1","2"]].values[270:]
y_train = df["W/L"].values[:270]
y_test = df["W/L"].values[270:]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(max_depth=2)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```
