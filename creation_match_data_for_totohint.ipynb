```python
df = pd.read_csv("./match_data_yearly/all_years.csv", index_col=0)
df = df[df["Year"] >= 2013]
df
```

```python
df=df[["Year","Sec","Date","Home","Away","HomeID","AwayID","Stadium"]]
df
```

```python
df_club_id = pd.read_csv("./club_and_id.csv")
df_club_id = df_club_id[:31]

for index, row in df.iterrows():
    home = row["Home"]
    away = row["Away"]
    
    for i,r in df_club_id.iterrows():
        if home == r["club"]:
            df.at[index,"Home"] = r["club_ja"]
            
        if away == r["club"]:
            df.at[index,"Away"] = r["club_ja"]
```
