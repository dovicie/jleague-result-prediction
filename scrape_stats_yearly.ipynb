```python
import re
import csv
import time
import requests
import datetime
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
year=2016
```

```python
url = f'https://www.football-lab.jp/shim/?year={year}' 
```

```python
r = requests.get(url)
r
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
records = soup.select('.statsTbl tr')
records
```

```python
data = []
```

```python
for record in records:
    l = [field.text for field in record.find_all(['th','td'])]           
    data.append(l)
```

```python
yearly = []
for d in data:
    d[1] = d[1].replace('\r','').replace('\t','').replace('\n','')
    yearly.append(d[1])
```

```python
yearly=yearly[1:]
```

```python
dev_rank = soup.select(f".seasonMenu  a[href='/shim/?year={year}'] .stnd")[0].string
```

```python
devision = int(dev_rank[1])
devision
```

```python
yearly.append(devision)
```

```python
yearly
```

```python
with open(f'./stats_yearly/shimizu-s-pluse/2018.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
