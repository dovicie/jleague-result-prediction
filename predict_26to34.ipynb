```python
from scipy.stats import poisson
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
df_AD = pd.read_csv("./match_data/HaHdAaAd.csv").set_index("team")
```

```python
df_AD
```

```python
df1_25 = pd.read_csv("./match_data/results_1to25.csv")
```

```python
df1_25
```

```python
aveHs = df1_25["Hs"].mean()
aveAg = df1_25["Ag"].mean()
```

```python
aveAg,aveHs
```

```python
df26_34 = pd.read_csv("./match_data/match_26to34.csv")
```

```python
df26_34
```

```python
x = np.arange(0,10)

for index,row in df26_34.iterrows():
    
    hscore = df_AD.at[row["ホーム"],"HA"] * df_AD.at[row["アウェイ"],"AD"] * aveHs
    ascore = df_AD.at["shim","HD"] * df_AD.at["kobe","AA"] * aveAg

    prediction_hs = np.argmax(poisson.pmf(k=x, mu=hscore))
    prediction_as =  np.argmax(poisson.pmf(k=x, mu=ascore))
    
    
    
    if prediction_hs > prediction_as:
        win_loss = 1
    elif prediction_hs < prediction_as:
        win_loss = 2
    else:
        win_loss = 0
        
    if row["ホーム得点"] == prediction_hs and row["アウェイ得点"] == prediction_as :
        errata = 2
    elif win_loss == row["勝敗"]:
        errata = 1
    else :
        errata = 0

#     print(f"{prediction_hs} - {prediction_as}",errata)
    with open('./match_data/predict_26to34.csv', 'a') as f:
        writer = csv.writer(f)
        writer.writerow([row["ホーム"], row["アウェイ"], prediction_hs, prediction_as, row["ホーム得点"], row["アウェイ得点"], errata])
```
