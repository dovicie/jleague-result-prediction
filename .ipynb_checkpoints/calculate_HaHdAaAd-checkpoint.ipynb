```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
df = pd.read_csv("./match_data/results_1to25.csv")
```

```python
df
```

```python
aveHs = df["Hs"].mean()
```

```python
aveHg = df["Hg"].mean()
```

```python
aveAs = df["As"].mean()
```

```python
aveAg = df["Ag"].mean()
```

```python
# HA = ホームでの攻撃力
# HD = ホームでの守備力
# AA = アウェーでの攻撃力
# AD = アウェーでの守備力

# 攻撃力は高いほど得点をとりやすいことを示しているのに対し、守備力は低いほど失点しにくいことを示している
# 例えばHA=1.5の場合、ホームで平均の1.5倍得点を取りやすく、HD=1.5ならば、ホームで平均の1.5倍得点を与えやすい

for index, row in df.iterrows():
    HA = row.Hs / aveHs
    HD = row.Hg / aveHg
    AA = row.As / aveAs
    AD = row.Ag / aveAg
    
    with open('./match_data/HaHdAaAd.csv', 'a') as f:
        writer = csv.writer(f)
        writer.writerow([row.team,HA,HD,AA,AD])
```
