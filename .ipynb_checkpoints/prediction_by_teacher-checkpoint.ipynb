```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('all_years.csv')
df
```

```python
df.columns
```

```python
df = df[['Year', 'Sec', 'Home', 'Away', 'W/L', 'Stadium', 'HomeElo', 'AwayElo', 'HomeED', 'AwayED']]
df
```

```python
df['W/L'] = df['W/L'].apply(lambda x: x if x != 2 else -1)
df
```

```python
df = pd.get_dummies(df)
df
```

```python
df_train = df[df['Year'].isin([2017, 2018, 2019])]
df_test = df[df['Year'] == 2020]
len(df_train), len(df_test)
```

```python
df_train['Year'] = df_train['Year'] - 2020
df_test['Year'] = df_test['Year'] - 2020
```

```python
columns = [column for column in df_train.columns if column != 'W/L']
X_train = df_train[columns].values
y_train = df_train['W/L'].values
X_test = df_test[columns].values
y_test = df_test['W/L'].values
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.manifold import TSNE

X_train_embedded = TSNE(random_state=1234).fit_transform(X_train)
```

```python
plt.scatter(X_train_embedded[:, 0], X_train_embedded[:, 1], c=y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = {'n_estimators': [100, 200, 300],
          'max_depth': [5, 6, 7, 8, 9, 10]}
clf = GridSearchCV(RandomForestClassifier(random_state=1234), params, cv=5, n_jobs=-1, verbose=1)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.score(X_test, y_test)
```

```python
pd.value_counts(df_train['W/L'])
```

```python
pd.value_counts(df_test['W/L'])
```

```python
len(clf.best_estimator_.feature_importances_), len(columns)
```

```python
feature_importances = sorted(zip(clf.best_estimator_.feature_importances_, columns), reverse=True)
```

```python
feature_importances[:10]
```

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
params = {'C': [0.01, 0.03, 0.1, 1, 3, 10],
          'gamma': [0.01, 0.03, 0.1, 1, 3, 10]}
clf = GridSearchCV(SVC(), params, cv=5, n_jobs=-1, verbose=1)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
df_mispredict = df_test[y_test != y_pred]
df_mispredict
```
