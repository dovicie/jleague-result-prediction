```python
import csv
import time
import requests
import datetime
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://data.j-league.or.jp/SFMS01/search?competition_years=2020&competition_frame_ids=1&competition_ids=477&competition_section_ids=4557&competition_section_ids=4558&competition_section_ids=4559&competition_section_ids=4560&competition_section_ids=4561&competition_section_ids=4562&competition_section_ids=4563&competition_section_ids=4564&competition_section_ids=4565&tv_relay_station_name='
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
records = soup.find_all('tr')
data = []
```

```python
for record in records:
    l = [field.text for field in record.find_all(['th','td'])]           
    data.append(l)
```

```python
with open(f'./match_data/match_26to34.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerows(data)
```
