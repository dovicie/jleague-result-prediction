```python
from scipy.stats import poisson
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
%matplotlib inline
```

```python
df_AD = pd.read_csv("./match_data/HaHdAaAd.csv").set_index("team")
df_1_25 = pd.read_csv("./match_data/results_1to25.csv")
df_26_34 = pd.read_csv("./match_data/match_26to34.csv")
```

```python
aveHs = df_1_25["Hs"].mean()
```

```python
hscores = df_AD.at["shim","HA"] * df_AD.at["kobe","AD"] * aveHs
```

```python
aveAg = df_1_25["Ag"].mean()
```

```python
ascores = df_AD.at["shim","HD"] * df_AD.at["kobe","AA"] * aveAg
```

```python
x = np.arange(10)
plt.plot(x, poisson.pmf(k=x,mu=hscores),"bo")
plt.vlines(x,0,poisson.pmf(k=x,mu=hscores))
```

```python
x = np.arange(10)
plt.plot(x, poisson.pmf(k=x,mu=ascores),"bo")
plt.vlines(x,0,poisson.pmf(k=x,mu=ascores))
```

```python
x = np.arange(0,10)
mu_list = [hscores,ascores]
for idx, mu in enumerate(mu_list):
    plt.plot(x, poisson.pmf(k=x, mu=mu), "o-", label='Ave scores={}'.format(round(mu,3)))
plt.legend()
plt.xlabel("scores")
plt.ylabel("probability")
```

```python
prediction_hs = np.argmax(poisson.pmf(k=x, mu=hscores))
```

```python
prediction_as =  np.argmax(poisson.pmf(k=x, mu=ascores))
```

```python
prediction = f"{prediction_hs} - {prediction_as}"
```
