```python
import csv
import time
import requests
import datetime
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("./match_data/match_26to34.csv") # csvを上書きしているので､すでにデータの前処理を行ったcsvである場合はエラーが出る｡
```

```python
df = df.replace("\n|\t", "", regex=True)
```

```python
df.dtypes
```

```python
import re

for index, row in df.iterrows():
    _ = re.sub('節第.日','', row["節"])
    _ = _.replace("第","")
    df.at[index, "節"] = _
```

```python
df["節"] = df["節"].astype(int)
```

```python
df
```

```python
df["試合日"] = df["試合日"].replace("\(.+\)","",regex = True)
```

```python
df
```

```python
f"{df['年度']} {df['試合日']} {df['K/O時刻']}"
```

```python
for index,row in df.iterrows():
    f"{row['年度']} {row['試合日']} {row['K/O時刻']}"
    _ =  pd.to_datetime(f"{row['年度']} {row['試合日']} {row['K/O時刻']}", format='%Y %m/%d %H:%M')
    df.at[index, "K/O時刻"] = _
```

```python
hscore_list = []
ascore_list = []
for index,row in df.iterrows():
    hscore,ascore = map(int,row["スコア"].split("-"))
    hscore_list.append(hscore)
    ascore_list.append(ascore)

    
```

```python
win_loss_list = []
for index,row in df.iterrows():
    if row["ホーム得点"] >row["アウェイ得点"]:
        win_loss_list.append(1)
    elif row["ホーム得点"] < row["アウェイ得点"]:
        win_loss_list.append(2)
    else:
        win_loss_list.append(0)

        
print(win_loss_list)
```
